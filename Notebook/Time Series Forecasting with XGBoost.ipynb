## Example for Recursive Forecasting

Recursive forecasting is a technique in time series forecasting where we predict multiple future steps by feeding each prediction back into the model as a new input for the next step. This approach works well with models like XGBoost, where you can use previous predictions as features for subsequent predictions.

Here’s a Python example of recursive forecasting using XGBoost:

```python
# Install XGBoost if not already installed
# !pip install xgboost

import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Generate synthetic time series data
np.random.seed(42)
time_series_length = 100
data = pd.DataFrame({
    'value': np.sin(np.linspace(0, 3 * np.pi, time_series_length)) + np.random.normal(scale=0.1, size=time_series_length)
})

# Create lag features for supervised learning
def create_lag_features(df, lag=3):
    for i in range(1, lag + 1):
        df[f'lag_{i}'] = df['value'].shift(i)
    df.dropna(inplace=True)
    return df

# Define the number of past time steps to use as features
lag = 3
data = create_lag_features(data, lag)

# Separate the training set
train_data = data[:-10]  # Using the last 10 points for recursive forecasting
X_train = train_data.drop('value', axis=1)
y_train = train_data['value']

# Initialize and train the XGBoost regressor
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, max_depth=3, learning_rate=0.1)
model.fit(X_train, y_train)

# Recursive Forecasting
n_forecasts = 10  # Number of future steps to forecast
last_values = data.drop('value', axis=1).iloc[-1].values  # Last known values to start forecasting
predictions = []

for i in range(n_forecasts):
    # Predict the next value
    pred = model.predict(last_values.reshape(1, -1))[0]
    predictions.append(pred)
    
    # Update the last_values array by removing the oldest lag and adding the new prediction
    last_values = np.roll(last_values, -1)
    last_values[-1] = pred

# Evaluate the model's forecasting accuracy on the test set
actual_values = data['value'].iloc[-n_forecasts:].values
mse = mean_squared_error(actual_values, predictions)
print("Mean Squared Error:", mse)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(range(len(data['value'])), data['value'], label='Actual Data', color='blue')
plt.plot(range(len(data['value']) - n_forecasts, len(data['value'])), predictions, label='Predictions', color='red')
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.title("Recursive Forecasting with XGBoost")
plt.legend()
plt.show()
```

### Explanation of the Code

1. **Lag Feature Creation**: The `create_lag_features` function generates lagged values as features. We use 3 lags (previous time steps) in this example, meaning each prediction is based on the last three time steps.

2. **Training the Model**: The model is trained on all available historical data except for the last 10 points, which are reserved for recursive forecasting.

3. **Recursive Forecasting Loop**:
   - We initialize `last_values` as the last set of lagged features from the training data.
   - In each iteration, we predict the next time step and store it in `predictions`.
   - We then update `last_values` by shifting values to the left and appending the new prediction at the end. This way, each prediction depends on previously predicted values.

4. **Evaluation and Plotting**: We calculate the Mean Squared Error (MSE) between the predicted and actual values for the 10-step forecast. Finally, we plot the actual and forecasted values to visualize the performance.

### Notes

- **Choosing the Number of Lags**: The number of lags affects the model's ability to capture the structure of the time series. Test different lag values to find the best fit for your data.
  
- **Stability**: Recursive forecasting can lead to error accumulation because each forecast depends on previous forecasts. Models can benefit from techniques like ensembling or more complex neural networks if stability becomes a problem.

This example should provide a foundation for using recursive forecasting with XGBoost. For more complex problems, combining recursive forecasting with other advanced techniques may yield better results.

````python
# Install XGBoost if not already installed
# !pip install xgboost

import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Generate synthetic time series data
np.random.seed(42)
time_series_length = 100
data = pd.DataFrame({
    'value': np.sin(np.linspace(0, 3 * np.pi, time_series_length)) + np.random.normal(scale=0.1, size=time_series_length)
})

# Create lag features for supervised learning
def create_lag_features(df, lag=3):
    for i in range(1, lag + 1):
        df[f'lag_{i}'] = df['value'].shift(i)
    df.dropna(inplace=True)
    return df

# Define the number of past time steps to use as features
lag = 3
data = create_lag_features(data, lag)

# Separate the training set
train_data = data[:-10]  # Using the last 10 points for recursive forecasting
X_train = train_data.drop('value', axis=1)
y_train = train_data['value']

# Initialize and train the XGBoost regressor
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, max_depth=3, learning_rate=0.1)
model.fit(X_train, y_train)

# Recursive Forecasting
n_forecasts = 10  # Number of future steps to forecast
last_values = data.drop('value', axis=1).iloc[-1].values  # Last known values to start forecasting
predictions = []

for i in range(n_forecasts):
    # Predict the next value
    pred = model.predict(last_values.reshape(1, -1))[0]
    predictions.append(pred)
    
    # Update the last_values array by removing the oldest lag and adding the new prediction
    last_values = np.roll(last_values, -1)
    last_values[-1] = pred

# Evaluate the model's forecasting accuracy on the test set
actual_values = data['value'].iloc[-n_forecasts:].values
mse = mean_squared_error(actual_values, predictions)
print("Mean Squared Error:", mse)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(range(len(data['value'])), data['value'], label='Actual Data', color='blue')
plt.plot(range(len(data['value']) - n_forecasts, len(data['value'])), predictions, label='Predictions', color='red')
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.title("Recursive Forecasting with XGBoost")
plt.legend()
plt.show()
````
